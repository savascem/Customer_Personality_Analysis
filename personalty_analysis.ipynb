```python
# fill NA
df["Income"] = df["Income"].fillna(df.groupby(["Education", "MntGoldProds"])["Income"].transform("mean"))
```

--------------------------------------------------------------------------------------------------------------------------

# TODO 3) Creating RFM Metrics

* already,we have the variable of Recency.
* frequency : NumWebPurchases + NumCatalogPurchases + NumStorePurchases

```python
# frequency 
df["frequency"] = df["NumWebPurchases"] + df["NumCatalogPurchases"] + df["NumStorePurchases"]
```

```python
# monetary
df["monetary"] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

--------------------------------------------------------------------------------------------------------------------------

# TODO 4) Calculate RF Score

```python
# R score
df["R_score"] = pd.qcut(df["Recency"], 5, labels=[5, 4, 3, 2, 1])

# F score
df["F_score"] = pd.qcut(df["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

# M score
df["M_score"] = pd.qcut(df["monetary"], 5, labels=[1, 2, 3, 4, 5])
```

--------------------------------------------------------------------------------------------------------------------------

# TODO 5) RF Segmentation

```python
# Creating RF_score
df["RF_score"] = df["R_score"].astype(str) + df["F_score"].astype(str)
```

```python
# segmentation map 
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

# creating segments
df["segment"] = df["RF_score"].replace(seg_map, regex=True)
```

```python
df[["RF_score", "segment"]].head(20)
```

# TODO 6) Analysis with the using RF segments

```python
df.groupby("segment").agg({"Income":["mean", "sum", "median", "std"]})
```

```python
# Creating new features and observing the variables that I chose in the pivot table

df["m/f"] = df["monetary"] / df["frequency"]
df["Age"] = 2015 - df["Year_Birth"]
df["Age_class"] = pd.cut(
                         df["Age"], 
                         [0, 18, 25, 35, 45, 55, 
                         df["Age"].max()], 
                         labels=["0_18", "19_25", "26_35", "36_45", "46_55", "56_max" ]
                         )
pd.pivot_table(
               df, 
               index=["segment", "Age_class"],
               values=["Income", "frequency", "Recency", "monetary", "m/f"], 
               #columns=["Complain"]
               aggfunc={
                        "Age_class": "count",
                        "Income": "mean", 
                        "frequency": "mean", 
                        "Recency": "mean", 
                        "monetary": ["mean", "std"],
                        "m/f": "mean",
                        },
               dropna=True,
               )
```
